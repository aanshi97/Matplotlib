# matplotlib

### line plot

```python
# importing matplotlib module

from matplotlib import pyplot as pt
```

```python
# x-axis valus

x = [5,2,9,4,7]
```

```python
# y-axis values

y = [10,5,8,4,2]
```

```python
# function to plot

pt.plot(x,y)
```

```python
# function to show the plot

pt.show()
```

### bar plot

```python
# function to plot bar

pt.bar(x,y)
pt.show()
```

```python
pt.show()
```

### histogram

```python
pt.hist(y)       # no any data is not repeat all data is only one time not any frequency for any number
```

```python
pt.hist(y)
pt.ylabel('frequency of numbers')
pt.xlabel('numbers (data)')
pt.show()
```

### scatter plot

```python
pt.scatter(x,y)
```

```python
pt.scatter([4,6,2,3],[9,6,3,10])
pt.show()
```

```python
pt.scatter([4,6,2,3],[9,6,3,10])
pt.axis([0,10,0,15])
pt.xlabel('x-axis')
pt.ylabel('y-axis')
pt.show()
```

### pie chart

```python
pt.pie(y)
pt.show()
```

```python
import numpy as np
t= np.arange(1,10,2)
t
```

```python
x1 = t
y1 = t

x2 = t
y2 = t**2

x3 = t
y3 = t**3

pt.plot(x1,y1,'r--')
```

```python
# draw multiple plots in a single graph.....here we plotng three plots

pt.plot(x1,y1,'r--', x2,y2,'bs', x3,y3,'g^')
pt.show()
```

```python
from matplotlib import pyplot as p
import numpy as n

x = n.arange(1,10,1)
y = n.arange(21,30,1)
p.scatter(x,y)
p.show()
```

```python
import matplotlib.pyplot as t
import numpy as a

x = a.arange(1,50,5)
y = a.arange(1,100,10)
t.scatter(x,y)
t.show()
```

```python
import matplotlib.pyplot as t
a = [2,10,7]
b = [8,3,5]

t.figure(figsize = (9,9))

t.subplot(221)
t.title('line plot')
t.plot(a,b)

t.subplot(222)
t.title('bar plot')
t.bar(a,b)

t.subplot(223)
t.title('scatter plot')
t.scatter(a,b)

t.subplot(224)
t.title('pie chart')
t.pie(a)

t.suptitle('categorical plotting')
t.show()

```
